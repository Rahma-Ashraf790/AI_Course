```python
def category_retirement_age(retirement_age):
    if 50 <= retirement_age <= 59:
        return 'Early'
    elif 60 <= retirement_age <= 65:
        return 'Normal'
    else:
        return 'Late'

user_df['retirement_Category'] = user_df['retirement_age'].apply(category_retirement_age)
```

```python
cards_dff = cards_df.groupby('client_id').agg({'card_brand' : 'count' , 'has_chip': 'max'}).reset_index()
cards_dff
```

```python
cards_dff['has_chip'].value_counts()
```

```python
total_user_card = pd.merge(user_df, cards_dff, on = 'client_id', how = 'left')
total_user_card
```

Data Summary

First user_df:
. There are around 2000 users in the dataset, while the cards dataset contains about 6000 cards. This means that most users have multiple cards
. retirement_ages are mostly between 60–65 years with no users below 50
. A large portion of users have a debt-to-income ratio > 1. This means that their debt exceeds their income
. Credit_Score ranges from 480 to 850 with a significant group achieving the maximum score of 850

Second cards_df:
. Most users have multiple cards
. There is Variety in card brands (Visa, Mastercard,..)
. Most cards have chips, but some users still use non-chip cards

Third Total_user_card:
. Most users with higher incomes have more cards
. Users with high credit_score (close to 850) have more or better cards
