42. Add Elements Conditionally

Write a NumPy program to add elements to a matrix. If an element in the matrix is 0, we will not add the element below this element.

````````````````````````````````````````python
array = np.array([[1, 1, 0, 2],
                  [0, 3, 0, 3],
                  [1, 0, 4, 4]])
print("Original array: ")
print(array)

print()
print("``````````````````")

res_sum = 0
for r in range(len(array)):
    for c in range(len(array[0])):
        if r < len(array) -1 and array[r][c] == 0:
            array[r+1][c] = 0
        res_sum += array[r][c]

print("Array after implementation: ")
print(array)

print()
print("``````````````````")

print(f"Sum is: {res_sum}")
                 
````````````````````````````````````````

````````````````````````````````````````python
arr = np.array([[[1,2,3,10],[4,5,6,3],[10,20,44,34]]])
print(f"Original array: ")
print(arr)

print()
print('````````````````````')

newarr = arr[::-1,::-1]
print(f"Array after swapping: ")
print(newarr)
````````````````````````````````````````

59. Element-Wise Multiply Two Arrays

Write a NumPy program to multiply two given arrays of the same size element-by-element.

````````````````````````````````````````python
first_arr = np.array([[2,3],[4,5]])
second_arr = np.array([[4,3],[2,8]])
result = np.multiply(first_arr, second_arr)
print(f"The multiplication array: ")
print(result)
````````````````````````````````````````
