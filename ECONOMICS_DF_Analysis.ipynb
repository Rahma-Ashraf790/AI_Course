```python
import pandas as pd
import numpy as np

df = pd.read_csv("world_economics.csv")
df 
```

- **borders** → الدول اللي لها حدود مشتركة مع الدولة 
- **GDP** → الناتج المحلي الإجمالي بمليار دولار  
- **GDP Growth** → معدل نمو الناتج المحلي % (قد إيه الاقتصاد بينمو)  
- **Interest Rate** → سعر الفائدة % (اللي بيحدده البنك المركزي)  
- **Inflation Rate** → معدل التضخم % (زيادة أسعار السلع والخدمات)  
- **Gov. Budget** → نسبة العجز أو الفائض في ميزانية الحكومة من الناتج المحلي % (سالب = عجز)  
- **Debt/GDP** → نسبة الدين العام للناتج المحلي % (مدى اعتماد الدولة على الديون)  
- **Current Account** → الحساب الجاري % من الناتج المحلي (يعبر عن الفرق بين الصادرات والواردات + التحويلات)

World Economics Data Summary

Dataset Overview:
. Contains 173 countries with 19 features (name, currency, GDP, population,...)
. Data includes geographic (region, borders) and economic metrics (GDP, inflation, debt, ...)
. Some columns (e.g., GDP Growth, Interest Rate) have missing values for smaller countries

Geographic Insights:
. Regions: Africa almost contains 50 countries, Asia, Americas, Europe, Oceania (least countries)
. English is the most common language, followed by Arabic, French, and Spanish
. Largest countries by area: Russia (17M km²), Canada (9.9M km²) ,China (9.7M km²), United States (9.4M km²)

Economic Insights:
. United States has highest GDP followed by China then Germany
. Sudan (272%) highest Debt/GDP ratios, followed by Japan (236.7%), Singapore (173.1%)
. Africa has the highest average GDP Growth among regions, with countries like Algeria (18.6%) standing out
. High Jobless rates in Africa (South Africa 33.2%) and some Asian subregions (West Asia 28.6%)

Financial Metrics:
. Inflation varies widely: Venezuela (172%) and Sudan (146.6%) have extreme rates, while Fiji (-0.4%) shows deflation
. Most countries have deficits (negative Gov. Budget) like United States (-6.4%), Tunisia (-6.6%)
. Current Account: Macau (31.4%) and Kuwait (28.2%) have strong surpluses, Lebanon (-31.5%) has a large deficit

Population Insights:
. Asia has maximum avg population (97.6M), Americas (36M), Africa (27.4M), Europe (18M), Oceania (8.9M)
. Top 5 populous countries: China, India , United States, Indonesia, Pakistan
. China and India greatly influence population distribution (Population outliers)
