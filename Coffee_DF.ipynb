```python
# add columns like sales_count, total_revenue and avg_price
co_df['Sales_Count'] = co_df['coffee_name'].map(co_df['coffee_name'].value_counts())
co_df['Total_Revenue'] = co_df['coffee_name'].map(co_df.groupby('coffee_name')['money'].sum())
co_df['Avg_Price'] = co_df['coffee_name'].map(co_df.groupby('coffee_name')['money'].mean())
```

Dataset Summary

Dataset Overview:
 . Contains 3547 rows with 11 features (coffee_name, money, cash_type,...)
 . Covers daily sales from early 2024 through 2025
 . Used to analyze product performance and time-based sales behavior

Product Insights:
 . The most sold coffee type is Latte
 . The most expensive product on average is Hot Chocolate
 . Price variation shows that higher price doesn’t always mean higher sales
 . Some products sell almost daily, while others are ordered less frequently

Time & Day Insights:
 . Afternoon is the peak period in both sales volume and revenue
 . Tuesday records the highest total sales among all weekdays
 . Sales on regular weekdays are higher than on the weekend
 . Sales are strongest between late morning and early afternoon hours

Monthly Insights:
 . March has the highest total sales in the dataset
 . A slight drop in sales is observed in the months following March
 . Comparing March 1, 2024 (396.3) vs March 1, 2025 (241.98) shows around a 39% decrease in revenue

Financial Insights:
 . Average transaction values remain relatively consistent despite price differences
 . No outlier or abnormal prices were detected
 . Afternoon contributes the most to total daily revenue
